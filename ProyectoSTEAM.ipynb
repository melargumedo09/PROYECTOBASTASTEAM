```python
var1=input("Dame una letra: ")
p1=0
p2=0
jugador1=input("Dame el nombre de jugador 1: ")
jugador2=input("Dame el nombre del jugador 2: ")

nombre1=input(jugador1 + " dame un nombre que empiece con " + var1 + ": ")
nombre2=input(jugador2 + " dame un nombre que empiece con " + var1 + ": ")

if(nombre1==nombre2):
  p1=p1+50
  p2=p2+50
else:
  p1=p1+100
  p2=p2+100

apellido1=input(jugador1 + " dame un apellido que empiece con: " + var1 + ": ")
apellido2=input(jugador2 + " dame un apellido que empiece con: " + var1 + ": ")

if(apellido1==apellido2):
  p1=p1+50
  p2=p2+50
else:
  p1=p1+100
  p2=p2+100

pais1=input(jugador1 + " dame un pais, estado o continente que empiece con: " + var1+ ": ")
pais2=input(jugador2 + " dame un pais, estado o continente que empiece con: " + var1+ ": ")

if(pais1==pais2):
  p1=p1+50
  p2=p2+50
else:
  p1=p1+100
  p2=p2+100

color1=input(jugador1 + " dame un color que empiece con: " + var1 + ": ")
color2=input(jugador2 + " dame un color que empiece con: " + var1 + ": ")

if(color1==color2):
  p1=p1+50
  p2=p2+50
else:
  p1=p1+100
  p2=p2+100

cosa1=input(jugador1 + " dame una cosa que empiece con: " + var1 + ": ")
cosa2=input(jugador2 + " dame una cosa que empiece con: " + var1 + ": ")

if(cosa1==cosa2):
  p1=p1+50
  p2=p2+50
else:
  p1=p1+100
  p2=p2+100

print("Puntaje final:")
print("El puntaje del jugador ", jugador1, " es: ", p1)
print("El puntaje del jugador ", jugador2, " es: ", p2)
```
